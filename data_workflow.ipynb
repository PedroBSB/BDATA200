  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>How Data Is Organized, Stored, and Accessed</em></p>

### **TABLES**
- Structured rows & columns
- Each row = one observation
- Each column = one variable
- Used in databases (SQL, Access)
- Enforces data types & constraints

### **SPREADSHEETS**
- Excel, Google Sheets
- Interactive cell-based editing
- Formulas & built-in functions
- Quick visualization tools
- Good for small-medium data

### **DATASETS**
- CSV, JSON, Parquet files
- Loaded into R, Python, etc.
- Scalable to millions of rows
- Version-controlled (Git)
- Best for reproducible analysis

  <p style="margin: 0; color: #4B2E83; font-weight: 600;">üí° Key Principle:</p>
  <p style="margin: 10px 0 0 0; color: #333;">Choose the storage format that best fits the size, complexity, and sharing needs of your data project.</p>
</div>

  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">üîÑ Reproducible Workflows</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>Ensuring Anyone Can Replicate Your Analysis</em></p>

### **WHAT IS REPRODUCIBILITY?**
- Same data + same code = same results
- Use scripts (R, Python) over manual steps
- Automate data cleaning pipelines
- Avoid copy-paste or point-and-click
- Share environment configs (packages, versions)

### **DOCUMENTING ANALYSIS STEPS**
- Record **why** decisions were made, not just **what**
- Use README files & inline comments
- Notebooks (Jupyter, R Markdown) blend code + narrative
- Document data sources & transformations
- Include metadata: date, author, version

<div style="background-color: #E8D3FF; border-left: 4px solid #4B2E83; padding: 15px 20px; margin: 20px 0; border-radius: 5px;">
  <p style="margin: 0; color: #4B2E83; font-weight: 600;">üí° Best Practice:</p>
  <p style="margin: 10px 0 0 0; color: #333;">Write your analysis as if a stranger will need to understand and rerun it six months from now.</p>
</div>

### **ERROR TRACKING**
- Log every error: what, when, why, how fixed
- Common sources: missing values, wrong data types, duplicates, outliers
- Use assertions & validation checks in code
- Maintain an error log (spreadsheet or issue tracker)
- Prevents repeating the same mistakes

### **PROCESS SHARING**
- Share code, data, and documentation together
- Use platforms: GitHub, GitLab, OSF
- Include clear instructions to re-run analysis
- Peer review catches errors early
- Collaboration improves quality & trust

  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>Core Terminology for Describing Data</em></p>
</div>

### **FEATURE**
- An input attribute used for prediction or modeling
- Common in machine learning
- **Example:** age, income, location in a predictive model

### **VARIABLE**
- Any measurable property that can take different values
- General statistics term
- **Example:** height, gender, test score

### **COVARIATE**
- A variable that may influence the outcome
- Controlled for in statistical models
- **Example:** age as a covariate in a drug trial

  <p style="margin: 10px 0 0 0; color: #333;">These terms often overlap. A "feature" in ML is typically called a "variable" in statistics and may serve as a "covariate" in regression analysis. Context determines the label.</p>

### **DATASET DICTIONARY**
A data dictionary documents every variable:
- Variable name & description
- Data type (numeric, text, date)
- Allowed values or range
- Units of measurement
- Missing value codes (NA, -999, blank)
- Source or collection method

### **DATA CLEANING BASICS**
Essential steps before any analysis:
- Handle missing values (remove, impute)
- Fix data type mismatches
- Remove duplicates
- Standardize formats (dates, names)
- Detect & handle outliers
- Validate against the data dictionary

### **DEFINING FUNCTIONS**
- Define with `def` keyword
- Can have default parameters
- Return values with `return`

````python
# Functions Example
def greet(name):
    return f"Hello, {name}!"

# Default parameters
def power(base, exp=2):
    return base ** exp

print(power(3))    # 9
print(power(3, 3)) # 27

# Data analysis function
def stats(data):
    avg = sum(data) / len(data)
    return {
        "mean": avg,
        "min": min(data),
        "max": max(data)
    }

print(stats([10, 20, 30]))
# {'mean': 20, 'min': 10, 'max': 30}
````

### **NUMPY BASICS**
- "Numerical Python" ‚Äî fast math on arrays
- Foundation of pandas, scikit-learn, etc.
- Core object: `ndarray` (N-dimensional array)

### **KEY OPERATIONS**
- Element-wise math operations
- Statistical functions
- Array shape manipulation

````python
import numpy as np

# Creating arrays
a = np.array([1, 2, 3])
b = np.zeros(5)           # [0,0,0,0,0]
c = np.arange(0, 10, 2)   # [0,2,4,6,8]
d = np.linspace(0, 1, 5)  # [0, 0.25, 0.5, 0.75, 1]

# Math (element-wise)
arr = np.array([10, 20, 30])
print(arr * 2)     # [20, 40, 60]
print(arr + 5)     # [15, 25, 35]

# Statistics
print(np.mean(arr))  # 20.0
print(np.std(arr))   # 8.16
print(np.max(arr))   # 30
print(arr.shape)     # (3,)
````

````python
import pandas as pd

# Series ‚Äî 1D labeled array
s = pd.Series([10, 20, 30], index=["a", "b", "c"])
print(s)

# DataFrame ‚Äî 2D table
df = pd.DataFrame({
    "name": ["A", "B", "C"],
    "score": [85, 92, 78]
})

print(df.head())       # first 5 rows
print(df.shape)        # (rows, cols)
print(df.dtypes)       # column types
print(df.describe())   # statistics

# Selecting a column
print(df["score"])        # Series
print(df["score"].mean()) # 85.0
````

### **ABOUT MATPLOTLIB**
- The most widely used Python plotting library
- Creates static publication-quality figures
- Works great in Jupyter/Colab
- Highly customizable (colors, labels, axes)

**Chart types:** Scatter, Line, Bar, Histogram, Box, Pie, Heatmap, and more

### **BASIC PATTERN**
```python
import matplotlib.pyplot as plt

# 1. Create figure
fig, ax = plt.subplots()

# 2. Plot data
ax.scatter(x, y)

# 3. Customize
ax.set_xlabel("X Label")
ax.set_ylabel("Y Label")
ax.set_title("Title")

# 4. Show
plt.show()
```

````python
import matplotlib.pyplot as plt
import pandas as pd

# Load data
url = "https://raw.githubusercontent.com/PedroBSB/BDATA200/refs/heads/main/Data/Auto.csv"
auto = pd.read_csv(url, na_values="?")

# Scatter Plot: MPG vs Weight
fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(auto["weight"], auto["mpg"], alpha=0.6, color="#4B2E83")
ax.set_xlabel("Weight", fontsize=12)
ax.set_ylabel("MPG", fontsize=12)
ax.set_title("MPG vs Weight", fontsize=14, fontweight="bold", color="#4B2E83")
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
````

````python
# Histogram: MPG Distribution
fig, ax = plt.subplots(figsize=(8, 5))
ax.hist(auto["mpg"], bins=20, color="#4B2E83", edgecolor="white", alpha=0.8)
ax.set_xlabel("MPG", fontsize=12)
ax.set_ylabel("Count", fontsize=12)
ax.set_title("MPG Distribution", fontsize=14, fontweight="bold", color="#4B2E83")
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
````

````python
# Bar Chart: Average MPG by Cylinders
avg = auto.groupby("cylinders")["mpg"].mean()

fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(avg.index.astype(str), avg.values, color="#4B2E83", alpha=0.8)
ax.set_xlabel("Cylinders", fontsize=12)
ax.set_ylabel("Average MPG", fontsize=12)
ax.set_title("Average MPG by Cylinders", fontsize=14, fontweight="bold", color="#4B2E83")
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
````

````python
# Line Plot: MPG Trends Over Years
avg_year = auto.groupby("year")["mpg"].mean()

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(avg_year.index, avg_year.values, marker="o", color="#4B2E83", linewidth=2, markersize=6)
ax.set_xlabel("Year", fontsize=12)
ax.set_ylabel("Average MPG", fontsize=12)
ax.set_title("MPG Trends Over Years", fontsize=14, fontweight="bold", color="#4B2E83")
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
````

### **ABOUT PLOTNINE**
- Python port of R's famous ggplot2
- Based on Grammar of Graphics ‚Äî build plots layer by layer
- Uses `+` to add layers (geoms, themes, scales)
- Beautiful defaults, easy themes

**Key geoms:** `geom_point`, `geom_histogram`, `geom_col`, `geom_line`, `geom_boxplot`, `geom_bar`

### **BASIC PATTERN**
```python
from plotnine import *

(ggplot(data, aes(x, y))
 + geom_point()    # layer
 + labs(title=...) # labels
 + theme_minimal() # theme
)
```

````python
from plotnine import *

# Scatter plot
(ggplot(auto, aes("weight", "mpg"))
 + geom_point(alpha=0.6, color="#4B2E83")
 + labs(title="MPG vs Weight", x="Weight", y="MPG")
 + theme_minimal()
)
````

````python
# Histogram
(ggplot(auto, aes("mpg"))
 + geom_histogram(bins=20, fill="#4B2E83", alpha=0.8)
 + labs(title="MPG Distribution", x="MPG", y="Count")
 + theme_bw()
)
````

````python
# Color by group
(ggplot(auto, aes("horsepower", "mpg", color="factor(cylinders)"))
 + geom_point(alpha=0.6)
 + labs(title="MPG vs Horsepower (by Cylinders)", x="Horsepower", y="MPG", color="Cylinders")
 + theme_minimal()
)
````

````python
# Box plot with fill
(ggplot(auto, aes("factor(cylinders)", "mpg", fill="factor(cylinders)"))
 + geom_boxplot(alpha=0.7)
 + labs(title="MPG by Cylinders", x="Cylinders", y="MPG", fill="Cylinders")
 + theme_bw()
)
````

### **ABOUT PLOTLY**
- Creates interactive charts (hover, zoom, pan)
- Works in Jupyter/Colab out of the box
- Uses `plotly.express` (high-level) or `graph_objects`
- Charts render as HTML ‚Äî great for sharing

**Website:** [https://plotly.com/python/](https://plotly.com/python/)

### **BASIC PATTERN**
```python
import plotly.express as px

# One-line charts!
fig = px.scatter(
    auto,
    x="weight",
    y="mpg",
    color="cylinders"
)
fig.show()
# Interactive in Colab!
```

````python
import plotly.express as px

# Interactive Scatter Plot
fig = px.scatter(
    auto,
    x="weight",
    y="mpg",
    color="cylinders",
    title="MPG vs Weight (Interactive)",
    labels={"weight": "Weight", "mpg": "MPG"},
    color_continuous_scale="Purples"
)
fig.show()
````

````python
# Interactive Histogram
fig = px.histogram(
    auto,
    x="mpg",
    nbins=20,
    title="MPG Distribution (Interactive)",
    labels={"mpg": "MPG"},
    color_discrete_sequence=["#4B2E83"]
)
fig.show()
````

````python
# Interactive Bar Chart
avg = auto.groupby("cylinders")["mpg"].mean().reset_index()

fig = px.bar(
    avg,
    x="cylinders",
    y="mpg",
    title="Average MPG by Cylinders (Interactive)",
    labels={"cylinders": "Cylinders", "mpg": "Average MPG"},
    color_discrete_sequence=["#4B2E83"]
)
fig.show()
````

````python
# Interactive Box Plot
fig = px.box(
    auto,
    x="cylinders",
    y="mpg",
    title="MPG by Cylinders (Interactive Box Plot)",
    labels={"cylinders": "Cylinders", "mpg": "MPG"},
    color="cylinders",
    color_discrete_sequence=px.colors.sequential.Purples
)
fig.show()
````
